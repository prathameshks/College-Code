```python
import pandas as pd
import seaborn as sns

# Load the Iris dataset
iris = sns.load_dataset('iris')

# For each feature
for feature in iris.columns[:-1]:  # Exclude the 'species' column
    Q1 = iris[feature].quantile(0.25)
    Q3 = iris[feature].quantile(0.75)
    IQR = Q3 - Q1

    # Define bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = iris[(iris[feature] < lower_bound) | (iris[feature] > upper_bound)]

    # Print outliers
    if not outliers.empty:
        print(f"Outliers for {feature} are:")
        print(outliers)
    else:
        print(f"No outliers found for {feature}.")
```
