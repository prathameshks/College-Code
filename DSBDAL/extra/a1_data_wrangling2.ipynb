```python
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.preprocessing import MinMaxScaler

# Create a DataFrame
df = pd.DataFrame({
    'Name': np.random.choice(['John', 'Anna', 'Peter', 'Linda', 'Mike'], 100),
    'Age': np.random.randint(10, 15, 100),
    'Subject': np.random.choice(['Math', 'Science', 'English', 'History', 'Geography'], 100),
    'Score': np.random.randint(50, 100, 100),
    'Height': np.random.normal(150, 10, 100),
    'Weight': np.random.normal(45, 5, 100)
})

# 1. Scan all variables for missing values and inconsistencies
missing_values = df.isnull().sum()
print("Missing values for each column:\n", missing_values)

# If there are missing values, you can use techniques like imputation to deal with them
# For example, fill missing values with the mean (for numerical columns)
for col in df.columns:
    if pd.api.types.is_numeric_dtype(df[col]):
        df[col] = df[col].fillna(df[col].mean())

# 2. Scan all numeric variables for outliers
# Here, we use Z-score to detect outliers
z_scores = np.abs(stats.zscore(df.select_dtypes(include=[np.number])))
outliers = (z_scores > 3).any(axis=1)
print("Outliers:\n", df[outliers])

# If there are outliers, you can use techniques like winsorization to deal with them
# For example, cap values at the 5th and 95th percentiles
for col in df.columns:
    if pd.api.types.is_numeric_dtype(df[col]):
        df[col] = np.clip(df[col], np.percentile(df[col], 5), np.percentile(df[col], 95))

# 3. Apply data transformations
# For example, apply Min-Max scaling on 'Age' column to change its scale
scaler = MinMaxScaler()
df['Age'] = scaler.fit_transform(df[['Age']])
```
