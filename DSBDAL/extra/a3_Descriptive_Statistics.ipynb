```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
```

```python
print("SD Of Age")
df['age'].std()
```

```python
# df.groupby('age')['charges'].agg(['mean','median','min','max','std'])

a = df.groupby('age')
mean = a['charges'].agg('mean')
median = a['charges'].agg('median')
min  = a['charges'].agg('min')
max = a['charges'].agg('max')
std = a['charges'].agg('std')

# merge with title in new dataframe
df2 = pd.concat([mean,median,min,max,std], axis=1)
df2.columns = ['mean','median','min','max','std']
df2
```
