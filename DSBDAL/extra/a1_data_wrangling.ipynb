```python
def get_info(column):
    print("No of Null Values:",column.isnull().sum())
    print("Mean:",column.mean())
    print("Median:",column.median())
    print("Mode:",column.mode())
    print("Standard Deviation:",column.std())
    print("Variance:",column.var())
    print("Skewness:",column.skew())
    print("Kurtosis:",column.kurt())
    print("Min:",column.min())
    print("Max:",column.max())
    print("Q1:",column.quantile(0.25))
    print("Q2:",column.quantile(0.50))
    print("Q3:",column.quantile(0.75))
    print("IQR:",column.quantile(0.75)-column.quantile(0.25))
    print("Range:",column.max()-column.min())
    print("**********")

def get_df_info(df):
    print("Total No of Rows:",df.shape[0])
    print("Total No of Columns:",df.shape[1])
    # print(df.dtypes)
    print(df.describe())
    print(df.isnull().sum())
```

```python
# fill with mean

get_info(df['Age'])
age = df['Age'].fillna(df['Age'].mean())
get_info(age)
```

```python
# fill with median

get_info(df['Age'])
age = df['Age'].fillna(df['Age'].median())
get_info(age)
```

```python
# fill with mode

get_info(df['Age'])
age = df['Age'].fillna(df['Age'].mode())
get_info(age)
```

```python
# delete rows with null values
dfnew = df
for i in dfnew.columns:
    dfnew = dfnew.dropna(subset=[i])

get_df_info(df)
get_df_info(dfnew)
```

```python
# last observation carried forward
dfnew = df
dfnew = dfnew.ffill()
get_df_info(df)
get_df_info(dfnew)
```

```python
# next observation carried backward
dfnew = df
dfnew = dfnew.bfill()
get_df_info(df)
get_df_info(dfnew)
```
