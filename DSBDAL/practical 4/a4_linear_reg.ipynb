```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load training data
train_data = pd.read_csv("train.csv")
```

```python
# Define features and target variable for training data
features = ["crim", "zn", "indus", "chas", "nox", "rm", "age", "dis", "rad", "tax", "ptratio", "lstat"]
target = "medv"
```

```python
# Visualize the distribution of predicted prices
plt.figure(figsize=(10, 6))
sns.histplot(train_data['medv'], kde=True, bins=30)
plt.title('Distribution of Train Prices')
plt.xlabel('Training Prices')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create and train the model
model = LinearRegression()
model.fit(train_data[features], train_data[target])
```

```python
# Load test data
test_data = pd.read_csv("test.csv")
```

```python
# Make predictions on test data
predictions = model.predict(test_data[features])
```

```python
# Create a DataFrame for the output
output = pd.DataFrame({'ID': test_data.ID, 'medv': predictions})
```

```python
# Visualize the distribution of predicted prices
plt.figure(figsize=(10, 6))
sns.histplot(output['medv'], kde=True, bins=30)
plt.title('Distribution of Predicted Prices')
plt.xlabel('Predicted Prices')
plt.ylabel('Frequency')
plt.show()
```

```python
# Write the output to a CSV file
output.to_csv('output.csv', index=False)
print("Predicted prices for test data have been saved to 'output.csv'.")
```
